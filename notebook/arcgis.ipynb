```python
import geopandas as gpd

# 读取GeoJSON文件
gdf = gpd.read_file("./arcgis/wind-raws-gj.json")

# 查看数据
gdf.head()
```

```python
gdf.describe()
```

```python
gdf.columns
```

```python
import plotly.express as px

# 使用plotly可视化GeoDataFrame
fig = px.scatter_mapbox(gdf, lat=gdf.geometry.y, lon=gdf.geometry.x, zoom=5, height=600, width=800)

# 设置图形标题
fig.update_layout(title_text="GeoDataFrame Visualization",mapbox_style="carto-positron")

# 显示图形
fig.show()
```
