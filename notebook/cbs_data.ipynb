```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 读取GeoJSON文件
gdf = gpd.read_file("./cbs/latest_cali_fires.geojson")

# 查看数据
gdf.head()
```

```python
gdf.describe()
```

```python
burned_area_km2 = round(gdf["acres_burned"].sum() * 0.00404686,2)
burned_area_km2
```

```python
# 使用matplotlib可视化GeoDataFrame
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax)

# 设置图形标题
plt.title("latest_cali_fires")

# 显示图形
plt.show()
```
