```python
import geopandas as gpd
import matplotlib.pyplot as plt

# 读取GeoJSON文件
gdf = gpd.read_file("./arcgis/wind-raws-gj.json")

# 查看数据
gdf.head()
```

```python
gdf.count()
```

```python
# 使用matplotlib可视化GeoDataFrame
fig, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax)

# 设置图形标题
plt.title("GeoDataFrame Visualization")

# 显示图形
plt.show()
```
